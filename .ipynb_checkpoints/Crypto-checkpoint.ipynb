```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bitcoin_price_df = pd.read_csv('Data/bitcoin_price.csv', parse_dates=['Date'])
bitcoin_price_df
```

```python
import matplotlib.dates as mdates
bitcoin_price_df['Date_mpl'] = bitcoin_price_df['Date'].apply(lambda x: mdates.date2num(x))

fig, ax = plt.subplots(figsize=(12,8))
sns.tsplot(bitcoin_price_df.Close.values, time=bitcoin_price_df.Date_mpl.values, ax=ax)
ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y.%m.%d'))
fig.autofmt_xdate()
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price in USD', fontsize=12)
plt.title("Closing price distribution of bitcoin", fontsize=15)
plt.show()
```
